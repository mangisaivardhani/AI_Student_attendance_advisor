```python
import pandas as pd

df = pd.read_csv("Student_Attendance.csv.csv")   # Replace with your file name

df.head()
```

```python
print(df.columns)
```

```python
df.info()

df.describe()
```

```python
def attendance_agent(attendance):

    if attendance >= 85:
        return "Excellent Attendance. Keep it up!"

    elif attendance >= 75:
        return "Attendance is good. Try to maintain it."

    else:
        return "Low Attendance! Attend extra classes and meet your faculty advisor."
```

```python
df["Agent Recommendation"] = df["Attendance"].apply(attendance_agent)

df.head()
```

```python
df.to_csv("Attendance_Advisor_Output.csv", index=False)

print("Output file created successfully.")
```
